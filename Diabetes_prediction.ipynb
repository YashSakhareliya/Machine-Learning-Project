## Work flow
-> collecting diabetes data \
-> data preprocessing \
-> train and test split data \
-> support vector machine classifier

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

collecting data


```python
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
# find number of rows and column
diabetes_dataset.shape
```

```python
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0-> non diabetes \
1-> diabetes

```python
diabetes_dataset.groupby('Outcome').mean()
```

spliting data into label wise x and y

```python
X = diabetes_dataset.drop(columns = 'Outcome', axis=1).reset_index(drop=True)
Y = diabetes_dataset['Outcome']
```

```python
X
```

```python
Y
```

Standerdize data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standerdize_data = scaler.transform(X)
```

```python
standerdize_data
```

```python
X = standerdize_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

spliting data train and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

training the model

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

model evalution

```python
# accuracy score on train data
x_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(x_train_prediction,Y_train)
```

```python
print("Accuracy score on training data: ", training_data_accuracy)
```

```python
# test data accuracy
x_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(x_test_prediction,Y_test)
```

```python
print("Accuracy score on testing data: ",test_data_accuracy)
```

# Making predictive syatem

```python
input_data = (6,148,72,35,0,33.6,0.627,50)

# chenging input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize input data
std_input_data = scaler.transform(input_data_reshaped)
print(std_input_data)

prediction = model.predict(std_input_data)
print(prediction)

if prediction[0] == 0:
  print("The person is not diabetic")
else:
  print("The person is diabetic")
```
